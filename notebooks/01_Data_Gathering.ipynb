#   Load Data

```python
dbutils.widgets.text("dataset", "Cars Datasets 2025.csv")
dataset = dbutils.widgets.get("dataset")
```

```python
# Load Cars Dataset 
file_path = f"/Volumes/workspace/default/ai-enhanced_project/{dataset}"

# Read CSV file into Spark DataFrame
df = spark.read.csv(file_path, header=True, inferSchema=True)

# Display basic information
display(df)
```

```python
# Show DataFrame schema and row count
print(f"Row count: {df.count()}")
df.printSchema()
```

```python
# Rename columns to remove invalid characters for Delta table
df = df.toDF(
    "company_names",
    "cars_names",
    "engines",
    "cc_battery_capacity",
    "horsepower",
    "total_speed",
    "performance_0_100_kmh",
    "cars_prices",
    "fuel_types",
    "seats",
    "torque"
)

# Export Spark DataFrame to Delta table
df.write.format("delta").mode("overwrite").saveAsTable("cars_dataset_delta")
```
